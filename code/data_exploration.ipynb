## Loading Data

```python
from utils import preprocess_dataset

train_dataset = preprocess_dataset("train", 100, 10)
test_dataset = preprocess_dataset("test", 100, 10)
valid_dataset = preprocess_dataset("valid", 100, 10)
```

```python
def get_number_feat_i(dataset, i):
    """
    nodes: i=0
    edges: i=1
    """
    return [dataset[j].stats[0, i].item() for j in range(len(dataset))]

train_nodes = get_number_feat_i(train_dataset, 0)
train_edges = get_number_feat_i(train_dataset, 1)
test_nodes = get_number_feat_i(test_dataset, 0)
test_edges = get_number_feat_i(test_dataset, 1)
valid_nodes = get_number_feat_i(valid_dataset, 0)
valid_edges = get_number_feat_i(valid_dataset, 1)
```

## Visualization

```python
import matplotlib.pyplot as plt

plt.hist(train_nodes, bins="auto", alpha=0.5, label='train')
plt.hist(test_nodes, bins="auto", alpha=0.5, label='test')
plt.hist(valid_nodes, bins="auto", alpha=0.5, label='valid')
plt.legend(loc='upper right')
plt.title("Number of nodes")
plt.show()
```

```python
plt.hist(train_edges, bins="auto", alpha=0.5, label='train')
plt.hist(test_edges, bins="auto", alpha=0.5, label='test')
plt.hist(valid_edges, bins="auto", alpha=0.5, label='valid')

plt.legend(loc='upper right')
plt.title("Number of edges")
plt.show()
```

```python
plt.scatter(train_nodes, train_edges, alpha=0.5, label='train')
plt.scatter(test_nodes, test_edges, alpha=0.5, label='test')
plt.scatter(valid_nodes, valid_edges, alpha=0.5, label='valid')

plt.legend(loc='upper right')
plt.title("Number of nodes vs number of edges")
plt.show()
```
