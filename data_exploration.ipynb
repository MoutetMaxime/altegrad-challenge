```python
from utils import preprocess_dataset_with_text
```

```python
d = preprocess_dataset_with_text("train", 50, 10)
```

```python
from torch_geometric.loader import DataLoader

train_loader = DataLoader(d, batch_size=256, shuffle=False)
```

```python
data = train_loader.dataset[9]
print(data)
```

```python
# load test data
d = preprocess_dataset_with_text("test", 50, 10)
test_loader = DataLoader(d, batch_size=256, shuffle=False)

data = test_loader.dataset[9]
print(data)
```

```python
# plot distribution of the number of nodes
nodes = [train_loader.dataset[i].stats[0, 0] for i in range(len(train_loader.dataset))]
edges = [train_loader.dataset[i].stats[0, 1] for i in range(len(train_loader.dataset))]

import matplotlib.pyplot as plt
import numpy as np

plt.figure()
plt.hist(nodes, bins=np.arange(0, 100, 1), alpha=0.5, label='nodes')
plt.hist(edges, bins=np.arange(0, 100, 1), alpha=0.5, label='edges')
plt.legend()
plt.show()

# plot data in a graph nodes vs edges
plt.figure()
plt.scatter(nodes, edges)

plt.xlabel("Number of nodes")
plt.ylabel("Number of edges")
plt.show()
```
